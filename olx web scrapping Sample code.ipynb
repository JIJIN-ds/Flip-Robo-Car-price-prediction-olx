```python
#Lets get data from all urls

num=0

for page in url:
    driver.get(page)
    time.sleep(10)
    num=num+1
    print(num," Url Executing")
    
    #Brand
    try:
        brand = driver.find_element_by_xpath("//div[@class='_35xN1']")
        Brand.append(brand.text)
    except:
        Brand.append("-")
    
    #Model
    try:
        model = driver.find_element_by_xpath("//div[@class='_35xN1']")
        Model.append(model.text.split()[-2])
    except:
        Model.append("-")
        
    #Variant
    try:
        variant = driver.find_element_by_xpath("//div[@class='_3tLee']")
        Variant.append(variant.text)
    except:
        Variant.append("-")
        
    #Year
    try:
        year = driver.find_element_by_xpath("//div[@class='_35xN1']")
        Year.append(year.text.split()[-1].replace("(",'').replace(")",''))
    except:
        Year.append("-")
        
    #fuel
    try:
        fuel = driver.find_element_by_xpath("//div[@class='aOxkz']/div[1]")
        Fuel.append(fuel.text)
    except:
        Fuel.append("-")

    #transmission
    try:
        transmission = driver.find_element_by_xpath("//div[@class='aOxkz']/div[3]")
        Transmission.append(transmission.text)
    except:
        Transmission.append("-")

    #km
    try:
        km = driver.find_element_by_xpath("//div[@class='aOxkz']/div[2]")
        kilometers.append(km.text)
    except:
        kilometers.append("-")
    
    #owner
    try:
        owner = driver.find_element_by_xpath("//div[@class='_1l939']/div[1]/div[2]/div[2]")
        owners.append(owner.text)
    except:
        owners.append("-")
    
    #location
    try:
        loc = driver.find_element_by_xpath("//div[@class='_1l939']/div[2]/div[2]/div[2]")
        location.append(loc.text.split()[-1])
    except:
        location.append("-")
        
    #price
    try:
        price = driver.find_element_by_xpath("//div[@class='_3FkyT']")
        Price.append(price.text.split()[1])
    except:
        Price.append("-")
```

```python
len(Brand)==len(Model)==len(Variant)==len(Year)==len(kilometers)==len(Fuel)==len(Transmission)==len(owners)==len(location)==len(Price)
```
